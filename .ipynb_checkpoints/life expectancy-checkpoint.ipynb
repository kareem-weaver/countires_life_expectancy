```python
pd.options.display.max_rows
pd.set_option('display.max_rows', None)
pd.pivot_table(df, index = ['Country', 'Year', 'Alcohol', 'BMI'], values = 'Life expectancy').sort_values('Year', ascending = False)
```

```python
for i in range(0,len(df)):
  df.pivot_table[i].Year = df.pivot_table[i].Year[:2014]
```

```python
# Set the maximum number of rows to display
pd.set_option('display.max_rows', None)

# Apply a filter to select rows from 2014
filtered_df = df[df['Year'] == 2014]

# Create the pivot table using the filtered DataFrame
pivot_table = pd.pivot_table(filtered_df, index=['Country', 'Year', 'Alcohol', 'BMI'], values='Life expectancy').sort_values('Year', ascending=False)

# Display the pivot table
pd.pivot_table(df, index = ['Country', 'Year', 'Alcohol', 'BMI'], values ='Life expectancy').sort_values('Year', ascending = False)
```
